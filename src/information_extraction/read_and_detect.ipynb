```python
from PIL import Image 
import pytesseract 
import sys 
from pdf2image import convert_from_path 
import os 
```

```python
PDF_file = "140-2020-014320.pdf"

# with the code from geeksforgeeks:
# https://www.geeksforgeeks.org/python-reading-contents-of-pdf-using-ocr-optical-character-recognition/

''' 
Part #1 : Converting PDF to images 
'''
  
# Store all the pages of the PDF in a variable 
pages = convert_from_path(PDF_file, 500) 
  
# Counter to store images of each page of PDF to image 
image_counter = 1
  
# Iterate through all the pages stored above 
for page in pages: 
  
    filename = "page_"+str(image_counter)+".jpg"
      
    # Save the image of the page in system 
    page.save(filename, 'JPEG') 
  
    # Increment the counter to update filename 
    image_counter = image_counter + 1
    
    ''' 
Part #2 - Recognizing text from the images using OCR 
'''
    
# Variable to get count of total number of pages 
filelimit = image_counter-1
  
# Creating a text file to write the output 
outfile = "out_text.txt"
  
# Open the file in append mode so that  
# All contents of all images are added to the same file 
f = open(outfile, "a") 
  
# Iterate from 1 to total number of pages 
for i in range(1, filelimit + 1): 

    filename = "page_"+str(i)+".jpg"
          
    # Recognize the text as string in image using pytesserct 
    text = str(((pytesseract.image_to_string(Image.open(filename), lang="fra")))) 
 
    text = text.replace('-\n', '')     
  
    # Finally, write the processed text to the file. 
    f.write(text) 
  
 # Close the file after writing all the text. 
f.close() 
```

```python
def language_detection(text):
    '''This function takes the text file and converts it a list of lines
    and detecting each line's language and store it in a new list of lines if it is in French.
    This fuction returns the list of lines in French'''
    
    import csv
    import spacy
    from spacy_langdetect import LanguageDetector

    file = open(text, 'r')
    reader = csv.reader(file)
    allRows = [row for row in reader]
    
    rows_in_fr = []
    nlp = spacy.load('en_core_web_sm')
    nlp.add_pipe(LanguageDetector(), name='language_detector', last=True)
    for row in allRows:
        doc = nlp(str(str(row)))
        if doc._.language["language"] == "fr" or row == [] or row[0].isnumeric():
            rows_in_fr.append(str(row))
    return (rows_in_fr)
```

```python
language_detection("out_text.txt")
```
